```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("../data/raw/creditcard.csv")
```

```python
df.shape
```

```python
df.head()
df.info()
df.describe()
```

```python
df["Class"].value_counts()
```

0 - Non-Fraudelent
1 - Fraudelent

```python
df["Class"].value_counts(normalize=True)
```

```python
fraud = df[df["Class"]==1]
normal = df[df["Class"]==0].sample(len(fraud), random_state=42)
```

```python
df["Class"].value_counts().plot(kind = "bar")
plt.title("Class Distribution")
plt.show()
```

```python
plt.hist(normal["Amount"], bins=50, alpha=0.5, label="Normal")
plt.hist(fraud["Amount"], bins=50, alpha=0.5, label="Fraud")
plt.legend()
plt.title("Amount: Fraud vs Normal Balanced")
plt.show()
```

```python
plt.hist(df["Time"], bins=50)
plt.title("Time Distribution")
plt.show()
```

```python
features = ["V1","V2","V3","V4"]
for f in features:
    plt.figure()
    plt.hist(df[df["Class"] == 0]["Amount"], bins=50, alpha=0.5, label="Normal")
    plt.hist(df[df["Class"] == 1]["Amount"], bins=50, alpha=0.5, label="Fraud")
    plt.legend()
    plt.title(f"{f} Distribution")
    plt.show()
    
```

```python
corr = df.corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

Certain transformed features show moderate correlation with fraud, suggesting predictive signal exist but distributed across multiple variables.
